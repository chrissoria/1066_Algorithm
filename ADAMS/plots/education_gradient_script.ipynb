```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

copying data from stata output

```python
def split_value_and_confidence(data):
    value, ci = re.split(r'[\(\)]', data)[:2]
    lower_ci, upper_ci = ci.split(', ')
    return value.strip(), lower_ci.strip(), upper_ci.strip()

data = {
    "Education Level": [],
    "Algorithm Name": [],
    "Margin": [],
    "Lower CI": [],
    "Upper CI": []
}

input_data = {
    "ADAMS": ["0.179 (0.134, 0.225)", "0.182 (0.117, 0.248)", "0.123 (0.073, 0.173)"],
    "Modified 10/66 (optimal)": ["0.325 (0.268, 0.381)", "0.240 (0.169, 0.312)", "0.131 (0.080, 0.182)"],
    "Modified 10/66 (ascribed)": ["0.266 (0.212, 0.319)", "0.164 (0.100, 0.228)", "0.110 (0.061, 0.158)"],
    "Langa-Weir": ["0.300 (0.242, 0.358)", "0.126 (0.068, 0.185)", "0.034 (0.005, 0.063)"],
    "Expert": ["0.301 (0.249, 0.352)", "0.289 (0.216, 0.362)", "0.143 (0.091, 0.195)"],
    "Hurd": ["0.296 (0.244, 0.349)", "0.271 (0.199, 0.343)", "0.164 (0.110, 0.218)"],
    "Lasso": ["0.374 (0.318, 0.430)", "0.303 (0.229, 0.379)", "0.177 (0.121, 0.233)"]
}

education_levels = ["Low education", "Medium education", "High education"]

for algorithm, values in input_data.items():
    for education_level, value in zip(education_levels, values):
        val, lower_ci, upper_ci = split_value_and_confidence(value)
        data["Education Level"].append(education_level)
        data["Algorithm Name"].append(algorithm)
        data["Margin"].append(val)
        data["Lower CI"].append(lower_ci)
        data["Upper CI"].append(upper_ci)

df = pd.DataFrame(data)
df
```

put data into proper format for plotting

```python
long_df = pd.DataFrame({
    "Education Level": data["Education Level"],
    "Algorithm Name": data["Algorithm Name"],
    "Margin": pd.to_numeric(data["Margin"]),
    "Lower CI": pd.to_numeric(data["Lower CI"]),
    "Upper CI": pd.to_numeric(data["Upper CI"])
})

long_df["Error Lower"] = long_df["Margin"] - long_df["Lower CI"]
long_df["Error Upper"] = long_df["Upper CI"] - long_df["Margin"]
```

```python
plt.rcParams.update({
    'font.size': 14,  # Set the global font size
    'axes.titlesize': 20,  # Set the font size of the title
    'axes.labelsize': 16,  # Set the font size of the x and y labels
    'xtick.labelsize': 14,  # Set the font size of the x-tick labels
    'ytick.labelsize': 16,  # Set the font size of the y-tick labels
    'legend.fontsize': 18,  # Set the font size of the legend
    'legend.title_fontsize': 16  # Set the font size of the legend title
})


color_map = {
    "Low education": "#D55E00",   # Vermilion
    "Medium education": "#0072B2", # Blue
    "High education": "#009E73"    # Green
}

fig, axes = plt.subplots(3, 3, figsize=(18, 15))

axes = axes.flatten()

for i, algorithm in enumerate(long_df["Algorithm Name"].unique()):
    ax = axes[i]
    subset = long_df[long_df["Algorithm Name"] == algorithm]
    # Plot error bars and points with color-blind friendly colors
    for level in subset["Education Level"].unique():
        level_data = subset[subset["Education Level"] == level]
        ax.errorbar(
            x=level_data["Education Level"], 
            y=level_data["Margin"], 
            yerr=[level_data["Error Lower"], level_data["Error Upper"]], 
            fmt='o', 
            c=color_map[level], 
            label=level,
            markersize=11
        )
    ax.set_ylim(0, 0.5) 
    ax.set_title(algorithm)
    ax.set_xlabel("")
    ax.set_ylabel("Margin")

# legend
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower right', ncol=3)

# drop empty subplots
for idx in range(len(long_df["Algorithm Name"].unique()), 9):
    fig.delaxes(axes[idx])

plt.tight_layout()
fig.savefig("education_gradients.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

cb_friendly_palette = ['#000000', '#FFB000', '#648FFF']

# Create a mapping between numeric race categories and their string representations
race_mapping = {1: 'White', 2: 'Black', 3: 'Hispanic'}

unique_races = probabilities['race_cat'].unique()
colors = {race: cb_friendly_palette[i % len(cb_friendly_palette)] for i, race in enumerate(unique_races)}

plt.figure(figsize=(12, 7))

# Scatter plot with colors based on race_cat
for race in unique_races:
    mask = probabilities['race_cat'] == race
    plt.scatter(probabilities.index[mask], 
                probabilities['k_fold_dem_pred_1066_av'][mask], 
                alpha=0.6, 
                c=[colors[race]], 
                label=race_mapping[race])  # Use the string representation here

plt.title('10/66 Algorithm Probability Scores with Ascribed Cutoff')
plt.xlabel('Person')
plt.ylabel('10/66 Probability Score')

plt.xticks([])

# whites
plt.axhline(y=0.08938869, color='#000000', linestyle='--', alpha=0.5, label='Whites 0.089')

# blacks
plt.axhline(y=0.25129401, color='#FFB000', linestyle='--', alpha=0.5, label='Blacks 0.252')

# hispanics
plt.axhline(y=0.88367432, color='#648FFF', linestyle='--', alpha=0.5, label='Hispanics 0.883')

# Add legend
plt.legend(title='Race and Cut Point', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()

plt.subplots_adjust(right=0.9)
plt.savefig("probabilities_rs_ct.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
probabilities['change'] = probabilities['k_fold_dem_pred_1066_av'] - probabilities['k_fold_dem_pred_1066_av_rs']
avg_change = probabilities.groupby('race_cat')['change'].mean()

cb_friendly_palette = ['#000000', '#FFB000', '#648FFF']
race_mapping = {1: 'White', 2: 'Black', 3: 'Hispanic'}

# order is White, Black, Hispanic
ordered_races = [1, 2, 3]
avg_change = avg_change.reindex(ordered_races)

plt.figure(figsize=(10, 6))

bars = plt.bar(range(len(avg_change)), avg_change.values, color=cb_friendly_palette)

plt.title('Average Change in 10/66 Algorithm Probability Scores by Race\n(Original - Race-Specific)')
plt.xlabel('Race')
plt.ylabel('Average Change in Probability Score')

plt.xticks(range(len(avg_change)), [race_mapping[race] for race in avg_change.index])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.4f}',
             ha='center', va='bottom')

# Add a horizontal line at y=0 for reference
plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig("probabilities_rs_probs_change.png", dpi=300, bbox_inches='tight')
plt.show()
```
