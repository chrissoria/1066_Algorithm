```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

copying data from stata output

```python
def split_value_and_confidence(data):
    value, ci = re.split(r'[\(\)]', data)[:2]
    lower_ci, upper_ci = ci.split(', ')
    return value.strip(), lower_ci.strip(), upper_ci.strip()

data = {
    "Education Level": [],
    "Algorithm Name": [],
    "Margin": [],
    "Lower CI": [],
    "Upper CI": []
}

input_data = {
    "ADAMS": ["0.179 (0.134, 0.225)", "0.182 (0.117, 0.248)", "0.123 (0.073, 0.173)"],
    "Modified 10/66 (optimal)": ["0.325 (0.268, 0.381)", "0.240 (0.169, 0.312)", "0.131 (0.080, 0.182)"],
    "Modified 10/66 (ascribed)": ["0.266 (0.212, 0.319)", "0.164 (0.100, 0.228)", "0.110 (0.061, 0.158)"],
    "Langa-Weir": ["0.300 (0.242, 0.358)", "0.126 (0.068, 0.185)", "0.034 (0.005, 0.063)"],
    "Expert": ["0.301 (0.249, 0.352)", "0.289 (0.216, 0.362)", "0.143 (0.091, 0.195)"],
    "Hurd": ["0.296 (0.244, 0.349)", "0.271 (0.199, 0.343)", "0.164 (0.110, 0.218)"],
    "Lasso": ["0.374 (0.318, 0.430)", "0.303 (0.229, 0.379)", "0.177 (0.121, 0.233)"]
}

education_levels = ["Low education", "Medium education", "High education"]

for algorithm, values in input_data.items():
    for education_level, value in zip(education_levels, values):
        val, lower_ci, upper_ci = split_value_and_confidence(value)
        data["Education Level"].append(education_level)
        data["Algorithm Name"].append(algorithm)
        data["Margin"].append(val)
        data["Lower CI"].append(lower_ci)
        data["Upper CI"].append(upper_ci)

df = pd.DataFrame(data)
df
```

put data into proper format for plotting

```python
long_df = pd.DataFrame({
    "Education Level": data["Education Level"],
    "Algorithm Name": data["Algorithm Name"],
    "Margin": pd.to_numeric(data["Margin"]),
    "Lower CI": pd.to_numeric(data["Lower CI"]),
    "Upper CI": pd.to_numeric(data["Upper CI"])
})

long_df["Error Lower"] = long_df["Margin"] - long_df["Lower CI"]
long_df["Error Upper"] = long_df["Upper CI"] - long_df["Margin"]
```

```python
color_map = {
    "Low education": "#D55E00",   # Vermilion
    "Medium education": "#0072B2", # Blue
    "High education": "#009E73"    # Green
}

fig, axes = plt.subplots(3, 3, figsize=(18, 15))

axes = axes.flatten()

for i, algorithm in enumerate(long_df["Algorithm Name"].unique()):
    ax = axes[i]
    subset = long_df[long_df["Algorithm Name"] == algorithm]
    # Plot error bars and points with color-blind friendly colors
    for level in subset["Education Level"].unique():
        level_data = subset[subset["Education Level"] == level]
        ax.errorbar(
            x=level_data["Education Level"], 
            y=level_data["Margin"], 
            yerr=[level_data["Error Lower"], level_data["Error Upper"]], 
            fmt='o', 
            c=color_map[level], 
            label=level
        )
    ax.set_title(algorithm)
    ax.set_xlabel("Education Level")
    ax.set_ylabel("Margin")

# legend
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', ncol=3)

# drop empty subplots
for idx in range(len(long_df["Algorithm Name"].unique()), 9):
    fig.delaxes(axes[idx])

plt.tight_layout()
fig.savefig("education_gradients.png", dpi=300)
plt.show()
```
